```python
import pandas as pd
from sklearn.model_selection import train_test_split
import ast
```

```python
class mydata():
    def __init__(self,path):
        data = pd.read_csv(path)
        data = data.drop(columns=['news_ID','hyperlink','content','html','clean_content'])
        X = data['full_content'].values.tolist()
        y = [ ast.literal_eval(i) for i in data['name'].values.tolist()]
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
path = './train.csv'
d = mydata(path)

for i in range(0,len(d.X_train)):
    if len(d.y_train[i]) is not 0:
        print(d.X_train[i])
        print(d.y_train[i])
```
