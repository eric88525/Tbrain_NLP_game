```python
import pandas as pd
from sklearn.model_selection import train_test_split
import ast
from opencc import OpenCC
import re
from zhon.hanzi import non_stops

class mydata():
    def __init__(self,path):
        data = pd.read_csv(path)
        data = data.drop(columns=['news_ID','hyperlink','content','html','clean_content'])
        X = [self.clean_string(i) for i in data['full_content'].values.tolist()]
        y = [ast.literal_eval(i) for i in data['name'].values.tolist()]
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X, y, test_size=0.1, random_state=42)
        
        self.train = []
        self.test = []
        cc = OpenCC('t2s')
        for i in range(0,len(self.X_train)):
            if str(self.X_train[i]) is not 'nan':
                if len(self.y_train[i]) is not 0:
                    start = []
                    end = []
                    for n in self.y_train[i]:                            
                        s = self.X_train[i].find(str(cc.convert(n)))
                        if s > 512:
                            continue
                        start.append(s)
                        end.append(s+len(str(n))-1)
                    self.train.append([self.X_train[i],self.y_train[i],start,end])
                else:
                    self.train.append([self.X_train[i],self.y_train[i],[],[]])
        for i in range(0,len(self.X_test)):
            if str(self.X_test[i]) is not 'nan':
                if len(self.y_test[i]) is not 0:
                    start = []
                    end = []
                    for n in self.y_test[i]:        
                        s = self.X_test[i].find(str(cc.convert(n)))
                        if s > 512:
                            continue
                        start.append(s)
                        end.append(s+len(str(n))-1)
                    self.test.append([self.X_test[i],self.y_test[i],start,end])
                else:
                    self.test.append([self.X_test[i],self.y_test[i],[],[]])
        """                           
        for i in range(0,len(self.X_test)):
            if self.X_test[i] is not 'nan':
                self.test.append([self.X_test[i],self.y_test[i]])
        """
        print(f'Train:{len(self.X_train)} Test:{len(self.X_test)}')
    def clean_string(self,content):
        if content is 'nan':
            return 'nan'
        cc = OpenCC('t2s')
        #cc = OpenCC('tw2sp')
        content = str(content).replace('\n','').replace('\t','').replace(' ','')
        content = re.sub("[\s+\.\!\/_,$%●▼►^*(+\"\']+|[+——！，。？、~@#￥%……&*（）]", "",content)
        content = re.sub(r"[%s]+" %non_stops, "",content)
        content = cc.convert(content)
        return content
        
```
