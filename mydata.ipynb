```python
import pandas as pd
from sklearn.model_selection import train_test_split
import ast
from opencc import OpenCC
import re
from zhon.hanzi import non_stops

class mydata():
    def __init__(self,path):
        data = pd.read_csv(path)
        data = data.drop(columns=['news_ID','hyperlink','content','html','clean_content'])
        X = [self.clean_string(i) for i in data['full_content'].values.tolist()]
        y = [ast.literal_eval(i) for i in data['name'].values.tolist()]
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X, y, test_size=0.1, random_state=42)
        
        self.train = []
        self.test = []
        for i in range(0,len(self.X_train)):
            if self.X_train[i] is not 'nan':
                self.train.append([self.X_train[i],self.y_train[i]])
        for i in range(0,len(self.X_test)):
            if self.X_test[i] is not 'nan':
                self.test.append([self.X_test[i],self.y_test[i]])
        print(f'Train:{len(self.X_train)} Test:{self.X_test}')
    def clean_string(self,content):
        cc = OpenCC('t2s')
        content = str(content).replace('\n','').replace('\t','').replace(' ','')
        content = re.sub("[\s+\.\!\/_,$%●▼►^*(+\"\']+|[+——！，。？、~@#￥%……&*（）]", "",content)
        content = re.sub(r"[%s]+" %non_stops, "",content)
        content = cc.convert(content)
        return content
        
```
